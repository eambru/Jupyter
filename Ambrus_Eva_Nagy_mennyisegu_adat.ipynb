```python
train = pd.DataFrame(train)

train['education'] = train['education'].replace([' Preschool', ' HS-grad', ' Doctorate'], 
                                            ['kindergarten', 'high school', 'PhD'])
train['education'] = train['education'].replace([' Some-college', ' Assoc-voc', ' Assoc-acdm', ' Bachelors'], 'Bsc')
train['education'] = train['education'].replace([' Masters', ' Professional school'], 'Msc')
train['education'] = train['education'].replace([' 1st',' 2nd',' 3rd',' 4th',' 5th',' 6th',' 7th',' 8th'], 'elementary school')
train['education'] = train['education'].replace([' 9th',' 10th',' 11th',' 12th'], 'secondary school')
train
```

```python
nameOfSchools = {
    'kindergarten':'Megtestesülés Óvoda',
    'elementary school':'Petőfi Sándor Általános Iskola',
    'secondary school':'Mechwart András Technikum',
    'high school':'Kossuth Gimnázium',
    'Bsc':'Semmelveis Egyetem',
    'Msc':'Debreceni Egyetem',
    'PhD':'Pázmány Péter Tudományegyetem'
}
```

```python
train['nameOfSchools'] = train['education'].map(nameOfSchools)
train
```

```python
nativeLanguage = {
    ' United-States' : 'English',
    ' Mexico' : 'Spanish',
    ' Philippines' : 'Filipino',
    ' Germany' : 'German',
    ' Canada' : 'English',
    ' Puerto-Rico' : 'Spanish',
    ' El-Salvador' : 'Spanish',
    ' India' : 'Hindi',
    ' Cuba' : 'Spanish',
    ' England' : 'English',
    ' Jamaica' : 'English',
    ' South' : 'Korean',
    ' China' : 'Chinese',
    ' Italy' : 'Italian',
    ' Dominican-Republic' : 'Spanish',
    ' Vietnam' : 'Vietnamese',
    ' Guatemala' : 'Spanish',
    ' Japan' : 'Japanese',
    ' Poland' : 'Polish',
    ' Columbia' : 'Spanish',
    ' Taiwan' : 'Chinese',
    ' Haiti' : 'French',
    ' Iran' : 'Arabic',
    ' Portugal' : 'Portuguese',
    ' Nicaragua' : 'Spanish',
    ' Peru' : 'Spanish',
    ' France' : 'French',
    ' Greece' : 'Greek',
    ' Ecuador' : 'Spanish',
    ' Ireland' : 'Irish',
    ' Hong' : 'Chinese',
    ' Cambodia' : 'Khmer',
    ' Trinadad&Tobago' : 'English',
    ' Laos' : 'Lao',
    ' Thailand' : 'Thai',
    ' Yugoslavia' : 'Slovenian',
    ' Outlying-US(Guam-USVI-etc)' : 'English',
    ' Honduras' : 'Spanish',
    ' Hungary' : 'Hungarian',
    ' Scotland' : 'English',
    ' Holand-Netherlands' : 'Dutch'
}
```

```python
train['nativeLanguage'] = train['native.country'].map(nativeLanguage)
train
```

```python
import random
languages = list(nativeLanguage.values())
train['learnedForeignLanguage'] = [random.choice(languages) for _ in range(len(train))]
levels = ['A','B','C']
subLevels = ['1','2']
train['levelOfSkill'] = [random.choice(levels)+random.choice(subLevels) for _ in range(len(train))]
train
```

```python
train.fillna(np.nan)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
y = train['target']
X = train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,
random_state = 1, stratify = y)
clf = RandomForestClassifier(n_estimators = 500, max_depth = 6)
clf.fit(X_train, y_train)
prediction = clf.predict(X_test)
from sklearn.metrics import accuracy_score
acc = accuracy_score(np.array(y_test), prediction)
acc
```
