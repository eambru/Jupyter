```python
#3. feladat
def morse(code):
  dicti = {	'.-'	:'A', 	'-.':'N',		'.----':'1',
	'-...':'B'	, 	'---':'O'	,'..---':'2',
	'-.-.':'C' ,	'.--.': 'P',	'...--':'3',
	'-..':'D',		'--.-':'Q'	, '....-':'4',
	'.':'E',	'.-.':'R'	,	'.....':'5',
	'..-.':'F',		'...':'S'	, 	'-....':'6',
	'--.':'G' ,	'-':'T'	,  '--...':'7',
	'....':'H' ,	'..-':'U',	'---..':'8',
	'..':'I',	'...-':'V',  '----.':'9',
	'.---':'J' ,	'.--':'W',	'-----':'0',
 '-.-':'K'	, '-..-':'X',
	'.-..':'L',	'-.--':'Y',
	'--':'M',	'--..':'Z'}
  decod=''
  for i in code.split(' '):
    if i == '/':
      decod +=' '
    else:
      decod += dicti.get(i,'')
  return decod
morse1= ".-- . .-.. -.-. --- -- . / .. -. / - .... . / -- .-.. / -.-. --- ..- .-. ... ."
d=morse(morse1)
print("A megoldás: ",d)
```
